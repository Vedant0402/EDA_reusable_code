# Day *2* Data Profiling & Summary Statistics


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.shape
```

```python
print(f"Number of rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
print("\nColumn Names:", df.columns.tolist())
```

### Data Types Summary

```python
df.dtypes
```

```python
print("Categorical Columns: ", df.select_dtypes(include="category").columns.tolist())
```

```python
print("Numeric Columns:", df.select_dtypes(include=np.number).columns.tolist())
```

```python
print("\nMissing Values:\n", df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
import missingno as msno
msno.matrix(df)
plt.show()
```

### Summary Statistics

```python
num_summary = df.describe().T
num_summary["missing_count"] = df.isnull().sum()
#num_summary["skewness"] = df.skew()
#num_summary["kurtosis"] = df.kurt()

display(num_summary)
```

```python
# Select only numeric columns
num_cols = df.select_dtypes(include=np.number).columns

# Create summary only for numeric features
num_summary = df[num_cols].describe().T

# Add skewness and kurtosis
num_summary["skewness"] = df[num_cols].skew()
num_summary["kurtosis"] = df[num_cols].kurt()

# Add missing value count (only for numeric columns)
num_summary["missing_count"] = df[num_cols].isnull().sum()

display(num_summary)
```

### 📘 Key Notes

Skewness: measures asymmetry (0 = normal, >0 right-skewed, <0 left-skewed)

Kurtosis: measures tail heaviness (>3 = heavy tails, <3 = light tails)

### Categorical Summary

```python
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    print(f"\nColumn: {col}")
    print(df[col].value_counts())
    print("-" * 40)
```

```python
plt.figure(figsize=(6,4))
for i, col in enumerate(cat_cols, 1):
    plt.subplot(1, len(cat_cols), i)
    sns.countplot(data=df, x=col)
    plt.title(col)
plt.tight_layout()
plt.show()
```

### Automated Data Profiling Function

```python
def data_profile(df):
    print("=== Dataset Summary ===")
    print(f"Shape: {df.shape}")
    print("\nData Types:\n", df.dtypes)
    print("\nMissing Values:\n", df.isnull().sum())
    print("\nDuplicate Rows:", df.duplicated().sum())

    print("\n=== Numerical Features Summary ===")
    display(df.describe().T)

    print("\n=== Categorical Features Unique Counts ===")
    for col in df.select_dtypes(include='object').columns:
        print(f"{col}: {df[col].nunique()} unique values")

data_profile(df)
```
